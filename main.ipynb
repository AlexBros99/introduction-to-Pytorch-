# Getting Start with Pytorch
### import libraries


```python
import torch
import torch.nn as nn 
import torch.optim as optim
import numpy as np
```

Creating Neural network, of define layers by using Sequential. <br>
By using torch.nn *libraries* to define the layer. <br>
Linear layer uses (1,1)parameters, indicates one feature *in* and one feature *out*. <br> 
From Neural networks, there is only one layer with one neuron in the entire neural network. 

```python
#Model 
model = nn.Sequential(nn.Linear(1,1))
```

For the optimizer with Pytorch has picked called sgd. <br>
Which stands for **stochastic gradient descent**. <br>
Of that will help to minmize the loss, bringing the formula closer <br>
to the correct answer.  

```python
#Loss and optimizer 
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

Storing the data in array under a TensorFlow framework '**tensor**'. 


```python
#Data
xs = torch.tensor([[-1.0],[0.0],[1.0],[2.0],[3.0],[4.0]]
                  ,dtype=torch.float32)
ys = torch.tensor([[-3.0],[-1.0],[1.0],[3.0],[5.0],[7.0]],
                  dtype=torch.float32)
```

Going through each line of the loop: <br>
1. "Zero the gradients" of the optimizer
2. Create an array of outputs that has been calculated the input **x** values. <br>
3. Compares the outputs with correct answers to calculate the loss (*value shows how bad or good the guess is*).<br>
4. Process called *backpropagattion* happens. Will give us the direction or gradient to give us the desired results. <br>
5. Updating the model parameters to the values based on gradients calculated in the **backpropagation** step. 

```python
#train
for _ in range(500):
    optimizer.zero_grad()
    outputs = model(xs)
    loss = criterion(outputs, ys)
    loss.backward()
    optimizer.step()
```

```python
#predict
with torch.no_grad():
    print(model(torch.tensor([[10.0]], dtype=torch.float32)))
```

```python
layer = model[0]
#Get the weights and bias
weights = layer.weight.data.numpy()
bias = layer.bias.data.numpy()
print("Weights: ", weights)
print("Bias: ", bias)
```
